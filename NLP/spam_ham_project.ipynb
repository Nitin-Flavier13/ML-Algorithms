```python
import numpy as np
import pandas as pd 
messages = pd.read_csv(r'C:\Users\Nitin Flavier\Desktop\Data Nexus\Data Science\ML_BootCamp\ML_Algos\NLP\Data\smsSpamCollection',sep='\t',names=['label','message'])
messages.head()
```

```python
# Data cleaning and processing 
import re
import nltk 
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords 
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer()

n = len(messages)
stop_words = stopwords.words('english')
```

```python
corpus = [] 
for i in range(n):
    message = re.sub('[^a-zA-Z]',' ',messages['message'][i])
    message = message.lower()
    message = [lemmatizer.lemmatize(w,pos='v') for w in message.split(' ') if (w != '') and (w not in stop_words)]
    message = ' '.join(message)
    corpus.append(message)
print(corpus)
```

### Bag of Words:


```python
y = np.where(messages['label'] == 'spam',1,0)
y
```

```python
# train-test split 
from sklearn.model_selection import train_test_split 
from sklearn.feature_extraction.text import CountVectorizer

X_train,X_test,y_train,y_test = train_test_split(corpus,y,test_size=0.33,random_state=32) 

cv = CountVectorizer(max_features=500,ngram_range=(1,2))

X_train = cv.fit_transform(X_train).toarray()
X_test = cv.transform(X_test).toarray()
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB 

nb = MultinomialNB()
model = nb.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print("Classification Report: ",classification_report(y_test,y_pred))
print()
print("Confusion Matrix: \n",confusion_matrix(y_test,y_pred))
print()
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

### Word2Vec

```python
from gensim.models import Word2Vec, KeyedVectors 
import gensim.downloader as api 

w2v = api.load('word2vec-google-news-300')
```

```python
messages.isna().sum()
```

```python
corpus = [] 
n = messages.shape[0]
for i in range(n):
    message = re.sub('[^a-zA-Z]',' ',messages['message'][i])
    message = message.lower()
    message = [lemmatizer.lemmatize(w,pos='v') for w in message.split(' ')]
    message = ' '.join(message)
    corpus.append(message)

print(messages.shape)
print(len(corpus))
```

```python
# representing every sentence from the word corpus
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

words = []
index_tobe_deleted = []

# in Word2Vec we represent each word in terms of feature_values, then for the sentence

for i,sent in enumerate(corpus):
    token = sent_tokenize(sent) # it will convert the sentence into words
    flag = 1
    for sent in token:
        flag = 0
        words.append(simple_preprocess(sent))
    
    if(flag):
        index_tobe_deleted.append(i)

print(index_tobe_deleted)
print(len(words))
print(words)
```

```python
import gensim

# lets train 
model = gensim.models.Word2Vec(words,epochs=500)
# vocabulary
model.wv.index_to_key
```

```python
print(model.corpus_count)
print(model.epochs)
```

```python
# to see similar words
model.wv.similar_by_word('kid')
```

```python
model.wv['good']
```

```python
def avgWord2Vec(sentence):

    return np.mean([model.wv[w] for w in sentence if w in model.wv.index_to_key], axis=0)
    
```

```python
from tqdm import tqdm

X = []

n = len(words)
y_index = []

for i in tqdm(range(n)):
    
    ele = avgWord2Vec(words[i])
    if not np.any(np.isnan(ele)):
        X.append(ele)
    else:
        y_index.append(i)

print(len(X))
print(len(y_index))
print(X[0][0:10])
```

```python
print("Input Size: ",len(X))
print("Feature Size: ",len(X[0])) 
```

```python
y = np.where(messages['label'] == 'spam',1,0)
y = np.delete(y,index_tobe_deleted)
y = np.delete(y,y_index)
```

```python
print(len(y))
len(X)
```

```python
# independent features:
df = pd.DataFrame()

for i in range(len(X)):
    tempdf = pd.DataFrame(X[i].reshape(1,-1))
    df = pd.concat([df,tempdf],ignore_index=True)

df.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df,y,test_size=0.33,random_state=32)
```

```python
from sklearn.ensemble import RandomForestClassifier 
rf = RandomForestClassifier()

rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)

print("Accuracy: ",accuracy_score(y_test,y_pred))
print()
print("Classification Report: ",classification_report(y_test,y_pred))
print()
print("Confusion Matrix: ",confusion_matrix(y_test,y_pred))
```
