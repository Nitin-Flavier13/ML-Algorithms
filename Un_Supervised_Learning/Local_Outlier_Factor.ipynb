```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X,y = make_blobs(n_samples=1000,centers=3,n_features=2,random_state=32)
```

```python
X
```

```python
X = np.vstack([X,[0,2.5]])
X = np.vstack([X,[2,0.0]])
X = np.vstack([X,[10,11.5]])
X = np.vstack([X,[8,1]])

y = np.append(y,3)
y = np.append(y,3)
y = np.append(y,3)
y = np.append(y,3)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.neighbors import LocalOutlierFactor

clf = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = clf.fit_predict(X)
```

```python
indexes = [i for i in range(len(y_pred)) if y_pred[i] == 1]
indexes = np.array(indexes)
out_indexes = [i for i in range(len(y_pred)) if y_pred[i] == -1]
out_indexes = np.array(out_indexes)
```

```python
plt.scatter(X[indexes,0],X[indexes,1],c=y[indexes])
```

```python
plt.scatter(X[out_indexes,0],X[out_indexes,1],c=y[out_indexes])
```
