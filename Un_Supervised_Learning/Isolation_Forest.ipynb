```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\Nitin Flavier\Desktop\Data Nexus\Data Science\ML_BootCamp\ML_Algos\Un_Supervised_Learning\datasets\expense.csv')
df.head()
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest

obj = IsolationForest(contamination=0.2)
obj.fit(df)
predictions = obj.predict(df)
predictions
```

```python
anomaly_indexes = np.where(predictions<0)[0]
df.iloc[anomaly_indexes]
```

```python
print(anomaly_indexes)
```

```python
filtered_data = df.drop(index=anomaly_indexes)
plt.scatter(filtered_data.iloc[:,0],filtered_data.iloc[:,1])
plt.scatter(df.iloc[anomaly_indexes,0],df.iloc[anomaly_indexes,1],edgecolors='r')
```
